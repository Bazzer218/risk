```python
def plot(ga):
    plot_args={'layout':(3, 8), 'figsize':(16,8)}
    ga.gene_df.drop('iteration', axis=1).hist(**plot_args)
    plt.show()
def run(ga, n):
    for i in range(n):
        ga.iteration()
        plot(ga)
    ga.save('risk_1.json')
    ga.save_log('risk_1.csv')
```
