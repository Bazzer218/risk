```python
def merit_fct(a):
    return len(a.internal()) + a.reinforcements()
```

```python
def best_choice(a, fct):
    retval = [ ]
    if len(a.border()) == 0: return None
    for opt in a.border():
        new_a = a.copy()
        new_a.add(opt)
        retval.append((new_a, opt, fct(new_a)))
    max_score = max([x[2] for x in retval])
    retval = [x for x in retval if x[2] == max_score]
    return random.sample(retval, 1)[0][1]
```

```python
def take_best(b, player, fct):
    region = best_choice(b.player(player), fct)
    if region is None: return
    own_re = b.player(player).reinforcements()
    opp_re = b.owner(region).reinforcements()
    chance = float(own_re) / (own_re+opp_re)
    if random.random() < chance:
        b.take(player, region)
```

```python
from multiprocessing import Pool
def multi(fct, n):
    p = Pool(7)
    retval = p.map(fct, list(range(n)))
    p.close()
    return retval
```

```python
def play_game(x, n_steps=299, fct=merit_fct):
    b = risk.Board(6)
    b.distribute()
    starting = b.copy()
    for x in range(n_steps):
        for i in range(b.n_players()):
            take_best(b, i, fct)
            if b.winner() is not None:
                return (starting, b.winner(), x)
    return (starting, None, None)
```

```python
%%time
results = multi(play_game, 1000)
```

```python
def in_continent(board, continent, player):
    if player is None: return 0
    a = board.player(player)
    return len([r for r in continent.regions() if r in a])
```

```python
def hist_plot(df, continent):
    x = df['random_in_'+continent]
    plt.hist(x, np.arange(min(x)-0.5, max(x)+1.5))
    x = df['winner_in_'+continent]
    plt.hist(x, np.arange(min(x)-0.5, max(x)+1.5), alpha=0.6)
    plt.title(continent)
    plt.legend(['random', 'winner'])
    plt.show()
```
